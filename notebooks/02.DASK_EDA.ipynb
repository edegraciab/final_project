```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar

traffic_accidents_raw = dd.read_csv('data/US_Accidents_March23.csv')
```

```python
traffic_accidents_raw.head()
```

```python
print(f"Rows: {traffic_accidents_raw.shape[0].compute()}, Columns: {traffic_accidents_raw.shape[1]}")  # Approximate row count
print(traffic_accidents_raw.dtypes)  # Data types
```

```python
missing_values = traffic_accidents_raw.isnull().sum()

with ProgressBar() as bar:
    porcent_missing = ( (missing_values / traffic_accidents_raw.size ) * 100).compute()
porcent_missing
```
